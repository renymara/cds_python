```python
import sys
import os
import pandas as pd
```

```python
# Adiciona o caminho da pasta 'utils' ao sys.path
sys.path.append(os.path.abspath(os.path.join('..', 'utils')))
import process_data as upd
```

```python
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df
```

```python
data_path = f"../data/zomato.csv"
df_original = pd.read_csv(data_path)
df = df_original.copy()
```

```python
def data_preparation(df):
    df.dropna(inplace = True)
    df.drop_duplicates(inplace = True)
    df = df.reset_index().drop(columns = 'index')
    df["Country Name"] = df["Country Code"].apply(upd.country_name)
    df['Price Tye'] = df['Price range'].apply(upd.create_price_tye)
    df['Color Name'] = df['Rating color'].apply(upd.color_name)
    df = upd.rename_columns(df)
    df['average_cost_for_two_brl'] = df.apply(upd.converter_para_brl, axis=1)
    df["cuisines"] = df.loc[:, "cuisines"].apply(lambda x: x.split(",")[0])
    return df
```

```python
df_processed = data_preparation(df)
```

```python
def export_processed_data(df):  
    df.to_csv("../data/zomato_processed.csv", index=False)
    return df
```

```python
export_processed_data(df_processed)
```
